```python
n_train = len(df_train)
all_data = pd.concat([df_train, df_test], axis=0).reset_index(drop=True)
```

```python
none_cols = [
    "PoolQC", "MiscFeature", "Alley", "Fence", "FireplaceQu",
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",
    "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2",
    "MasVnrType", "MSSubClass"
]
for col in none_cols:
    if col in all_data.columns:
        all_data[col] = all_data[col].fillna("None")
```

```python
zero_cols = [
    "GarageYrBlt", "GarageArea", "GarageCars",
    "BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF", "TotalBsmtSF",
    "BsmtFullBath", "BsmtHalfBath", "MasVnrArea"
]
for col in zero_cols:
    if col in all_data.columns:
        all_data[col] = all_data[col].fillna(0)
```

```python
all_data["LotFrontage"] = (
    all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
)


for col in all_data.select_dtypes(include="object").columns:
    all_data[col] = all_data[col].fillna(all_data[col].mode()[0])


for col in all_data.select_dtypes(include=np.number).columns:
    all_data[col] = all_data[col].fillna(all_data[col].median())

print(f"Missing after fill: {all_data.isnull().sum().sum()}")
```

```python
all_data["TotalSF"]       = all_data["TotalBsmtSF"] + all_data["1stFlrSF"] + all_data["2ndFlrSF"]
all_data["TotalBath"]     = (all_data["FullBath"] + 0.5 * all_data["HalfBath"] +
                              all_data["BsmtFullBath"] + 0.5 * all_data["BsmtHalfBath"])
all_data["TotalPorch"]    = (all_data["OpenPorchSF"] + all_data["EnclosedPorch"] +
                              all_data["3SsnPorch"] + all_data["ScreenPorch"] +
                              all_data["WoodDeckSF"])
all_data["HouseAge"]      = all_data["YrSold"] - all_data["YearBuilt"]
all_data["RemodAge"]      = all_data["YrSold"] - all_data["YearRemodAdd"]
all_data["IsRemodeled"]   = (all_data["YearBuilt"] != all_data["YearRemodAdd"]).astype(int)
all_data["HasGarage"]     = (all_data["GarageArea"] > 0).astype(int)
all_data["HasPool"]       = (all_data["PoolArea"] > 0).astype(int)
all_data["HasFireplace"]  = (all_data["Fireplaces"] > 0).astype(int)
all_data["HasBasement"]   = (all_data["TotalBsmtSF"] > 0).astype(int)
all_data["Qual_SF"]       = all_data["OverallQual"] * all_data["TotalSF"]

print("New features added: TotalSF, TotalBath, TotalPorch, HouseAge, RemodAge, "
      "IsRemodeled, HasGarage, HasPool, HasFireplace, HasBasement, Qual_SF")
```

```python
ordinal_map = {
    "ExterQual":  {"Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "ExterCond":  {"Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "BsmtQual":   {"None": 0, "Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "BsmtCond":   {"None": 0, "Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "BsmtExposure": {"None": 0, "No": 1, "Mn": 2, "Av": 3, "Gd": 4},
    "BsmtFinType1": {"None": 0, "Unf": 1, "LwQ": 2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6},
    "BsmtFinType2": {"None": 0, "Unf": 1, "LwQ": 2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6},
    "HeatingQC":  {"Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "KitchenQual":{"Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "FireplaceQu":{"None": 0, "Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "GarageFinish":{"None": 0, "Unf": 1, "RFn": 2, "Fin": 3},
    "GarageQual": {"None": 0, "Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "GarageCond": {"None": 0, "Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5},
    "PoolQC":     {"None": 0, "Fa": 1, "TA": 2, "Gd": 3, "Ex": 4},
    "Fence":      {"None": 0, "MnWw": 1, "GdWo": 2, "MnPrv": 3, "GdPrv": 4},
    "LandSlope":  {"Sev": 1, "Mod": 2, "Gtl": 3},
    "LotShape":   {"IR3": 1, "IR2": 2, "IR1": 3, "Reg": 4},
    "PavedDrive": {"N": 0, "P": 1, "Y": 2},
    "CentralAir": {"N": 0, "Y": 1},
    "Functional": {"Sal": 1, "Sev": 2, "Maj2": 3, "Maj1": 4,
                   "Mod": 5, "Min2": 6, "Min1": 7, "Typ": 8},
}
```

```python
for col, mapping in ordinal_map.items():
    if col in all_data.columns:
        all_data[col] = all_data[col].map(mapping).fillna(0).astype(int)

# 3-d. One-hot encode remaining categoricals
cat_cols = all_data.select_dtypes(include="object").columns.tolist()
all_data = pd.get_dummies(all_data, columns=cat_cols, drop_first=True)
print(f"Shape after one-hot encoding: {all_data.shape}")

# 3-e. Fix skewed numeric features (log1p transform)
num_feats = all_data.select_dtypes(include=np.number).columns
skewness  = all_data[num_feats].apply(lambda x: skew(x.dropna()))
skewed    = skewness[skewness.abs() > 0.75].index.tolist()
all_data[skewed] = np.log1p(all_data[skewed].clip(lower=0))
print(f"Log1p applied to {len(skewed)} skewed features")

# 3-f. Log-transform the target
y_log = np.log1p(y)

# 3-g. Split back
X_train = all_data.iloc[:n_train].copy()
X_test  = all_data.iloc[n_train:].copy()

print(f"\nFinal train shape : {X_train.shape}")
print(f"Final test  shape : {X_test.shape}")
```

```python
scaler = RobustScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)



N_FOLDS = 5
kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=42)

def rmsle_cv(model, X, y, scaled=True):
    """Return mean ± std RMSLE via KFold CV."""
    Xdata = X_train_sc if scaled else X_train.values
    scores = -cross_val_score(model, Xdata, y,
                               scoring="neg_root_mean_squared_error",
                               cv=kf, n_jobs=-1)
    return scores.mean(), scores.std()

results = {}
```

```python
ridge = Ridge(alpha=10, random_state=42)
lasso = Lasso(alpha=0.0005, max_iter=10000, random_state=42)
enet  = ElasticNet(alpha=0.0005, l1_ratio=0.9, max_iter=10000, random_state=42)

for name, model in [("Ridge", ridge), ("Lasso", lasso), ("ElasticNet", enet)]:
    m, s = rmsle_cv(model, X_train_sc, y_log)
    results[name] = m
    print(f"{name:15s}  RMSLE = {m:.5f}  ±{s:.5f}")

# 5-b. Tree-based models (no scaling needed)
rf  = RandomForestRegressor(n_estimators=300, max_features="sqrt",
                             n_jobs=-1, random_state=42)
gbm = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                 max_depth=4, max_features="sqrt",
                                 min_samples_leaf=15, min_samples_split=10,
                                 loss="huber", random_state=42)

for name, model in [("RandomForest", rf), ("GradientBoosting", gbm)]:
    m, s = rmsle_cv(model, X_train, y_log, scaled=False)
    results[name] = m
    print(f"{name:15s}  RMSLE = {m:.5f}  ±{s:.5f}")

if XGB_AVAILABLE:
    xgb_model = xgb.XGBRegressor(
        colsample_bytree=0.4603, gamma=0.0468, learning_rate=0.05,
        max_depth=3, min_child_weight=1.7817, n_estimators=2200,
        reg_alpha=0.4640, reg_lambda=0.8571, subsample=0.5213,
        silent=True, random_state=42, n_jobs=-1,
        eval_metric="rmse", verbosity=0
    )
    m, s = rmsle_cv(xgb_model, X_train, y_log, scaled=False)
    results["XGBoost"] = m
    print(f"{'XGBoost':15s}  RMSLE = {m:.5f}  ±{s:.5f}")

if LGB_AVAILABLE:
    lgb_model = lgb.LGBMRegressor(
        objective="regression", num_leaves=5, learning_rate=0.05,
        n_estimators=720, max_bin=55, bagging_fraction=0.8,
        bagging_freq=5, feature_fraction=0.23, feature_fraction_seed=9,
        bagging_seed=9, min_data_in_leaf=6, min_sum_hessian_in_leaf=11,
        random_state=42, n_jobs=-1, verbose=-1
    )
    m, s = rmsle_cv(lgb_model, X_train, y_log, scaled=False)
    results["LightGBM"] = m
    print(f"{'LightGBM':15s}  RMSLE = {m:.5f}  ±{s:.5f}")
```

```python

base_estimators = [
    ("ridge",  Ridge(alpha=10, random_state=42)),
    ("lasso",  Lasso(alpha=0.0005, max_iter=10000, random_state=42)),
    ("gbm",    GradientBoostingRegressor(
                   n_estimators=1000, learning_rate=0.05, max_depth=4,
                   max_features="sqrt", min_samples_leaf=15,
                   min_samples_split=10, loss="huber", random_state=42)),
]
if XGB_AVAILABLE:
    base_estimators.append(("xgb", xgb.XGBRegressor(
        n_estimators=500, learning_rate=0.05, max_depth=3,
        colsample_bytree=0.5, subsample=0.5, random_state=42,
        eval_metric="rmse", verbosity=0, n_jobs=-1)))
if LGB_AVAILABLE:
    base_estimators.append(("lgb", lgb.LGBMRegressor(
        n_estimators=300, learning_rate=0.05, num_leaves=5,
        feature_fraction=0.23, bagging_fraction=0.8, bagging_freq=5,
        random_state=42, n_jobs=-1, verbose=-1)))

stacker = StackingRegressor(
    estimators=base_estimators,
    final_estimator=Ridge(alpha=1.0),
    cv=5, n_jobs=-1
)

m, s = rmsle_cv(stacker, X_train, y_log, scaled=False)
results["Stacking"] = m
print(f"{'Stacking':15s}  RMSLE = {m:.5f}  ±{s:.5f}")

# ── 7. COMPARE MODELS ─────────────────────────────────────────────────────────
plt.figure(figsize=(10, 5))
pd.Series(results).sort_values().plot(kind="barh", color="steelblue")
plt.xlabel("CV RMSLE (lower is better)")
plt.title("Model Comparison – Cross-Validation RMSLE")
plt.tight_layout()
plt.savefig("05_model_comparison.png", dpi=120)
plt.close()
print("\nModel comparison chart saved.")
```

```python

best_name  = min(results, key=results.get)
print(f"Best model: {best_name}  (RMSLE = {results[best_name]:.5f})")


final_model = stacker
final_model.fit(X_train, y_log)
preds_log   = final_model.predict(X_test)
preds       = np.expm1(preds_log)            # undo log1p
preds       = np.clip(preds, 0, None)        # safety: no negative prices


train_preds_log = final_model.predict(X_train)
train_preds     = np.expm1(train_preds_log)

rmse_train = np.sqrt(mean_squared_error(y, train_preds))
mae_train  = mean_absolute_error(y, train_preds)
r2_train   = r2_score(y, train_preds)
print(f"\nTrain-set diagnostics (overfitting check):")
print(f"  RMSE : ${rmse_train:,.0f}")
print(f"  MAE  : ${mae_train:,.0f}")
print(f"  R²   : {r2_train:.4f}")


residuals = y - train_preds
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].scatter(train_preds, residuals, alpha=0.4, s=15)
axes[0].axhline(0, color="red", lw=1)
axes[0].set_xlabel("Predicted SalePrice")
axes[0].set_ylabel("Residuals")
axes[0].set_title("Residuals vs Predicted")

stats.probplot(residuals, plot=axes[1])
axes[1].set_title("Q-Q Plot of Residuals")
plt.tight_layout()
plt.savefig("06_residuals.png", dpi=120)
plt.close()

# Actual vs Predicted
plt.figure(figsize=(7, 7))
plt.scatter(y, train_preds, alpha=0.4, s=15)
lims = [min(y.min(), train_preds.min()), max(y.max(), train_preds.max())]
plt.plot(lims, lims, "r--", lw=1)
plt.xlabel("Actual SalePrice")
plt.ylabel("Predicted SalePrice")
plt.title("Actual vs Predicted (Train Set)")
plt.tight_layout()
plt.savefig("07_actual_vs_predicted.png", dpi=120)
plt.close()

```
